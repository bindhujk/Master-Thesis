```python
import numpy as np
import pandas as pd

# Function to create train/val/test splits based on time series order
def make_splits(df, id_cols, date_col="date",
                frac_train=0.7, frac_val=0.15):
    # Sort by id_cols and date_col to ensure temporal order within each series
    df = df.sort_values(id_cols + [date_col]).copy()
    df["split"] = np.nan
    # Assign splits for each group
    for key, grp in df.groupby(id_cols):
        idx = grp.index.values
        n   = len(idx)
        # Determine split indices
        t1  = int(frac_train * n)
        t2  = t1 + int(frac_val   * n)
        # Assign splits
        df.loc[idx[:t1],  "split"] = "train"
        df.loc[idx[t1:t2], "split"] = "val"
        df.loc[idx[t2:],  "split"] = "test"
    return df

m5 = pd.read_csv("C:\\Users\\Bindhu\\Documents\\!Thesis\\m5_final\\Data\\m5_preprocessed_fe.csv", parse_dates=["date"])
m5 = make_splits(m5, id_cols=["id"])
m5.to_csv("C:\\Users\\Bindhu\\Documents\\!Thesis\\m5_final\\Data\\m5_final.csv", index=False)
```
